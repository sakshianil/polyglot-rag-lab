# 01 — Multimodal Vector Database

This notebook demonstrates the first stage of a multimodal RAG pipeline: loading sample files, converting them into text records, chunking them, and preparing them for embedding and vector database storage.


## Pipeline

Raw files → parsing → text extraction → chunking → embeddings → vector database.


```python
from pathlib import Path

DATA_DIR = Path('../data/raw')
for path in DATA_DIR.rglob('*'):
    if path.is_file():
        print(path)
```

```python
def simple_chunks(text, chunk_size=500, overlap=50):
    chunks = []
    start = 0
    while start < len(text):
        end = start + chunk_size
        chunks.append(text[start:end])
        start = end - overlap
        if start < 0:
            start = 0
    return chunks
```
