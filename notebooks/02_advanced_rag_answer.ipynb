# 02 — Advanced RAG Answer Generation

This notebook shows how retrieved chunks can be converted into a structured answer prompt.


```python
question = 'Was sind typische deutsche Gerichte? Nenne mir bitte auch ein Rezept vegan.'
retrieved_context = '''
Typical German dishes include Kartoffelsalat, Sauerkraut, Bratwurst, Schnitzel, Spätzle, Sauerbraten, Eintopf, and Brezeln.
A vegan Kartoffelsalat can be prepared with potatoes, onion, vegetable broth, vinegar, mustard, plant oil, salt, pepper, and herbs.
'''

prompt = f'''
You are a helpful RAG assistant. Answer in the user's language.
Use the retrieved context. If context is insufficient, say so clearly.

Question:
{question}

Retrieved context:
{retrieved_context}

Answer in a structured and readable format.
'''
print(prompt)
```
