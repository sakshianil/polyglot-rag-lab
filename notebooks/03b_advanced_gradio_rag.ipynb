# 03b — Advanced Gradio RAG

This notebook outlines an advanced Gradio RAG interface with retrieval preview, answer controls, and multilingual behavior.


```python
import gradio as gr

def advanced_answer(question, top_k=3, language='auto'):
    retrieved = [
        'Typical German dishes include Kartoffelsalat, Sauerkraut, Bratwurst, Schnitzel and Spätzle.',
        'A vegan Kartoffelsalat can be prepared with potatoes, vegetable broth, vinegar, mustard, oil and herbs.'
    ]
    answer = 'Structured RAG answer generated from retrieved context.'
    return '\n'.join(retrieved), answer

with gr.Blocks() as demo:
    gr.Markdown('# Advanced Gradio RAG')
    question = gr.Textbox(label='Your Question')
    top_k = gr.Slider(1, 10, value=3, step=1, label='Top-k Retrieved Chunks')
    language = gr.Dropdown(['auto', 'English', 'German'], value='auto', label='Answer Language')
    ask = gr.Button('Ask Question')
    retrieved = gr.Textbox(label='Retrieved Context')
    answer = gr.Textbox(label='AI Response')
    ask.click(advanced_answer, inputs=[question, top_k, language], outputs=[retrieved, answer])

# demo.launch()
```
